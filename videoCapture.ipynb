```python
####show recording video from a directory
import cv2
import numpy as np
import time
cap=cv2.VideoCapture('videos/videoCapture.mp4')
if cap.isOpened==False:
    print('ERROR FILE NOT FOUND')
while(True):
    ret,frame=cap.read()
    if ret==True:
        #write 20 FPS --file per second
        time.sleep(1/20)
        cv2.imshow("capture video",frame)
        if cv2.waitKey(10) & 0XFF == ord('q'):
            break 
    else:
        break
cap.release()
cv2.destroyAllWindows()
```

```python
####capture and show video from camera

import cv2
import numpy as np
cap=cv2.VideoCapture(0) ### 0 means use default camera from device

width=int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height=int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

##winows --*'DIVX'
##macOs or Linux --*'XVID'

# fourcc=cv2.VideoWriter_fourcc(*'DIVX')
videoCap=cv2.VideoWriter('videos/videoCapture.mp4',cv2.VideoWriter_fourcc(*'DIVX'),20,(width,height))

while(True):
    ret,frame=cap.read()
   ###capture video
    grayFrame=cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)
    cv2.imshow("input RGB",frame)
    cv2.imshow('gray image',grayFrame)
    
    ##DRAWING IMAGE
    videoCap.write(frame)
    ##escape 
    if cv2.waitKey(1) & 0XFF == ord('q'):
        break
cap.release()
videoCap.release()
cv2.destroyAllWindows()
```

```python
###Drawing a shape on video stream
import cv2
import numpy as np

cap=cv2.VideoCapture(0) ##0 means use default web cam camera
width=int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height=int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

##TOP LEFT CONRDER
x=width//2
y=height//2

##width and height of rectangle
w=width//4
h=height//4

##BOTTOM RIGHT CORNER OF rectangle x+w,y+h

while(True):
    ret,frame=cap.read()
#     cv2.line()
    cv2.rectangle(frame,(x,y),(x+w,y+h),color=(255,0,21),thickness=5)
    cv2.imshow('VideoRactangle',frame)
    if cv2.waitKey(1) & 0XFF == ord('q'):
        break
    
cap.release()    
cv2.destroyAllWindows()
```

```python
##drawing shape using mouse
import cv2
import numpy as np

########Functions########
def draw_circle(event,x,y,flags,param):
    global pt1,pt2,topLeft_click,bottomRight_click
    if event == cv2.EVENT_LBUTTONDOWN:  ####event for left button
        if topLeft_click==True and bottomRight_click==True:
            pt1=(0,0)
            pt2=(0,0)
            topLeft_click=False
            bottomRight_click=False
        if topLeft_click==False:
            pt1=(x,y)
            topLeft_click=True
        elif bottomRight_click==False:
            pt2=(x,y)
            bottomRight_click=True
            
        

#####        
##declare a global variable
pt1=(0,0)
pt2=(0,0)
topLeft_click=False
bottomRight_click=False

#####capture a video
cap=cv2.VideoCapture(0)

cv2.namedWindow(winname='Image')
cv2.setMouseCallback('Image',draw_circle)

while(True):
    ret,frame=cap.read()
    if topLeft_click:
        cv2.circle(frame,pt1,2,color=(0,0,255),thickness=-1)
    if topLeft_click and bottomRight_click:
        cv2.rectangle(frame,pt1,pt2,color=(0,0,255),thickness=2)
    cv2.imshow("Image",frame)
    
    if cv2.waitKey(1) & 0XFF ==ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
