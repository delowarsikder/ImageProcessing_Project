```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def display(img,cmap='gray'):
    fig=plt.figure(figsize=(10,12))
    ax=fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
```

```python
inputImg=cv2.imread('images/ratingContest.jpg')
inputImg=cv2.cvtColor(inputImg,cv2.COLOR_BGR2RGB)
# display(inputImg)
```

```python
featureImg=cv2.imread("images/yesNo.jpg")
featureImg=cv2.cvtColor(featureImg,cv2.COLOR_BGR2RGB)
# display(featureImg)
```

```python
orb=cv2.ORB_create()
```

```python
kp1,dst1=orb.detectAndCompute(inputImg,None)##return key point and descriptor two parameter
kp2,dst2=orb.detectAndCompute(featureImg,None)
```

```python
##applying a bruteforce technique
##this process will good fit if two images most of same feature
bf=cv2.BFMatcher(cv2.NORM_HAMMING,crossCheck=True)
```

```python
matches=bf.match(dst1,dst2)
```

```python
single_match=matches[0]
```

```python
single_match.distance
```

```python
len(matches)
```

```python
matches=sorted(matches,key=lambda x:x.distance)
```

```python
featureMatch=cv2.drawMatches(inputImg,kp1,featureImg,kp2,matches,None,flags=2)
```
