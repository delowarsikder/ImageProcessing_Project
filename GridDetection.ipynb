```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
chessBoard_img=cv2.imread('images/chessBoard.png')
chessBoard_img=cv2.cvtColor(chessBoard_img,cv2.COLOR_BGR2GRAY)
plt.imshow(chessBoard_img,cmap='gray')
```

```python
found,corner=cv2.findChessboardCorners(chessBoard_img,(7,7))
```

```python
found
```

```python
# corner
cv2.drawChessboardCorners(chessBoard_img,(7,7),corner,found)
```

```python
plt.imshow(chessBoard_img,cmap='gray')
```
