```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
cvlogo=cv2.imread('images/opencvlogo.png')
plt.imshow(cvlogo)
cvlogo.shape
```

```python
logo_gray=cv2.cvtColor(cvlogo,cv2.COLOR_BGR2GRAY)
plt.imshow(logo_gray,cmap='gray')
logo_gray.shape
```

```python
##bit wise not operation
##white convert into black and vice-versa
msk_logo=cv2.bitwise_not(logo_gray)
plt.imshow(msk_logo,cmap='gray')
msk_logo.shape
```

```python
##to create white backgound
import numpy as np
white_background=np.full(msk_logo.shape,255,dtype=np.uint8)
white_background=cv2.cvtColor(white_background,cv2.COLOR_BGR2RGB)
# plt.imshow(white_background)
# white_background.shape
```

```python
background=cv2.bitwise_or(white_background,white_background,mask=msk_logo)
plt.imshow(background)
# cv2.imwrite('images/cvBlack.png',background)
```
