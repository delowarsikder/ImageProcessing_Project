```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
###load images
def load_img():
    img=cv2.imread('images/suduko.png')
    img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    return img
```

```python
### display image
def disply_img(img):
    fig=plt.figure(figsize=(5,5))
    ax=fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
```

```python
routine_img=cv2.imread("images/routine.jpg",cv2.IMREAD_GRAYSCALE)
suduko_img=cv2.imread("images/suduko.png",cv2.IMREAD_GRAYSCALE)
disply_img(suduko_img)
```

```python
##sobol x calculation
sobolX=cv2.Sobel(suduko_img,cv2.CV_64F,1,0,ksize=5)
disply_img(sobolX)
```

```python
###sobol y calculation
sobolY=cv2.Sobel(suduko_img,cv2.CV_64F,0,1,ksize=5)
disply_img(sobolY)
```

```python
###use laplacial opetator
laplacian_img=cv2.Laplacian(suduko_img,cv2.CV_64F,ksize=5)
disply_img(laplacian_img)
```

```python
laplacian_img=cv2.Laplacian(suduko_img,cv2.CV_64F)
disply_img(laplacian_img)
```

```python
##blening two image
alpha=0.5
beta=0.5
gamma=0
blendingXY=cv2.addWeighted(sobolX,alpha,sobolY,beta,gamma)
disply_img(blendingXY)
```

```python
###check thresh hold
ret,th1=cv2.threshold(suduko_img,127,suduko_img.max(),cv2.THRESH_BINARY)
disply_img(th1)
```

```python
###create a kernel
kernel=np.ones(shape=(4,4),dtype=np.uint8)
```

```python
##using morphology gradiant
gradient=cv2.morphologyEx(blendingXY,cv2.MORPH_GRADIENT,kernel)
disply_img(gradient)
```
