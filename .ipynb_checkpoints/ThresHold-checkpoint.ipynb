```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img=cv2.imread("images/cover.png")
img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(img,cmap='gray')
```

```python
###threshold
ret,thresh1=cv2.threshold(img,127,img.max(),cv2.THRESH_BINARY)
```

```python
print(ret)
plt.imshow(thresh1,cmap='gray')
```

```python
#####inverse threshold
ret1,thres2=cv2.threshold(img,127,255,cv2.THRESH_BINARY_INV)
plt.imshow(thres2,cmap='gray')
```

```python
##Adaptive threshold
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
text=cv2.imread("images/text.jpg",cv2.IMREAD_GRAYSCALE)
plt.imshow(text,cmap='gray')
textCp=text.copy()
```

```python
ret2,th2=cv2.threshold(text,127,255,cv2.THRESH_BINARY_INV)
plt.imshow(th2,cmap='gray')
```

```python
ret2,th2=cv2.threshold(text,127,255,cv2.THRESH_BINARY)
plt.imshow(th2,cmap='gray')
```

```python
def show_pic(img):
    fig=plt.figure(figsize=(20,20))
    ax=fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
```

```python
# show_pic(text)
```

```python
###adptive thres  ,,for this process image must in convert into gray scale
from matplotlib import pyplot as plt
th3=cv2.adaptiveThreshold(text,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,11,8)
show_pic(th3)
show_pic(textCp)
```

```python
##gamma correction
bleng_img=cv2.addWeighted(src1=th3,alpha=0.4,src2=text,beta=0.6,gamma=0)
show_pic(bleng_img)
```
