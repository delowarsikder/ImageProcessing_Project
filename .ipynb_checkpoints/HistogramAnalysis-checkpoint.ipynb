```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fix_img=cv2.imread('images/messi.jpg')
messi_img=cv2.cvtColor(fix_img,cv2.COLOR_BGR2RGB)
plt.imshow(messi_img)
# display_img(messi_img)
```

```python
messi_img.shape
```

```python
##display function
def display_img(img):
    fig=plt.figure(figsize=(1,5))
    ax=fig.add_subplot(111)
    ax.imshow(img)
```

```python
###Histogram plot
hist_val=cv2.calcHist([messi_img],channels=[0],mask=None,histSize=[256],ranges=[0,256])
plt.plot(hist_val)  ##this is not a image just plot a graph
```

```python
##to show all color histogram
color=('b','g','r')
for i,col in enumerate(color):
    histr=cv2.calcHist([messi_img],[i],None,[256],[0,256])
    plt.plot(histr,color=col)
    plt.xlim([0,250])
#     plt.ylim([0,1000]) ###set for y axis
plt.title("Histogram For a messi")    
```

```python
###hsto gramm equlization
img=cv2.imread("images/leaf.jpg")
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
img.shape
```

```python
img.shape[:2]
```

```python
mask=np.zeros(shape=img.shape[:2],dtype=np.uint8)
plt.imshow(mask,cmap='gray')
```

```python
##some portion also black
mask_img=mask
mask_img[100:150,100:200]=255
plt.imshow(mask_img,cmap='gray')
```

```python
##masking
mask_img=cv2.bitwise_and(mask_img,mask_img,mask)
plt.imshow(mask,cmap='gray')
```

```python
##masking with a input image
mask_input_img=cv2.bitwise_and(img,img,mask=mask)
plt.imshow(mask_input_img)
```
