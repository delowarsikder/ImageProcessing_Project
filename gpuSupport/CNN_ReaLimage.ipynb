```python
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
```

```python
##dataPath
path="E:/Machine_Learning_A_Z/CNN/dataset"
```

```python
cat=cv2.imread('E:/Machine_Learning_A_Z/CNN/dataset/test/cats/cat.4001.jpg')
cat=cv2.cvtColor(cat,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(cat,cmap='gray')
```

```python
cat.shape
```

```python
from keras.preprocessing.image import ImageDataGenerator
```

```python
image_gen=ImageDataGenerator(rotation_range=30,
                             width_shift_range=0.1,
                             height_shift_range=0.1,
                            rescale=1/255,
                            shear_range=0.2,
                            zoom_range=0.2,
                            horizontal_flip=True,
                            fill_mode='nearest')
```

```python
plt.imshow(image_gen.random_transform(cat),cmap='gray')
```

```python
image_gen.flow_from_directory(path+'/train')
```

```python
image_gen.flow_from_directory(path+'/test')
```

```python
##import library
from keras.models import Sequential
from keras.layers import Conv2D,Flatten,Dense,MaxPooling2D,Activation,Dropout
```

```python
model=Sequential()
```

```python
##create a model
##layer-1
model.add(Conv2D(filters=32,kernel_size=(3,3),input_shape=(64,64,3),activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))

#layer-2
model.add(Conv2D(filters=64,kernel_size=(3,3),input_shape=(64,64,3),activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))

#layer-3
model.add(Conv2D(filters=64,kernel_size=(3,3),input_shape=(64,64,3),activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))

##add flatten
model.add(Flatten())

###adding dense layer
model.add(Dense(128))
model.add(Activation('relu'))
model.add(Dropout(0.5))  ##selecting 50% neuron from total nueron to prevent overfit

model.add(Dense(1))## 1 for binary outcome 0->cat,1->dog
model.add(Activation('sigmoid'))

model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.summary()
```

```python
train_img_gen=image_gen.flow_from_directory(path+'/train',target_size=(64,64),class_mode='binary',batch_size=32)
```

```python
# from PIL import Image
```

```python
test_img_gen=image_gen.flow_from_directory(path+'/test',target_size=(64,64),class_mode='binary',batch_size=32)
```

```python
train_img_gen.class_indices
```

```python
##step_per_epoch=total_train/batch_size
results=model.fit_generator(train_img_gen,steps_per_epoch=250,validation_data=test_img_gen,epochs=1000,validation_steps=63)
```

```python
##to ignore warnig
import warnings
warnings.filterwarnings('ignore')
```

```python
results.history['accuracy']
```

```python
plt.plot(results.history['accuracy'])
```
