```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
path='G:/Study/3-2/Project_3-2/systemDevelopment/chest_xray'
```

```python
img=cv2.imread(path+'/test/NORMAL/IM-0030-0001.jpeg')
plt.imshow(img)
```

```python
from keras.preprocessing.image import ImageDataGenerator
```

```python
image_gen=ImageDataGenerator(rotation_range=30,
                             width_shift_range=0.1,
                             height_shift_range=0.1,
                             zoom_range=0.2,
                             shear_range=0.2,
                             rescale=1/255,
                             horizontal_flip=True,
                             fill_mode='nearest',)
```

```python
plt.imshow(image_gen.random_transform(img))
```

```python
## Apply image generator all image
```

```python
path
```

```python
image_gen.flow_from_directory(path+'/train')
```

```python
image_gen.flow_from_directory(path+'/validation')
```

```python
##create a model
```

```python
from keras.models import Sequential
from keras.layers import Conv2D,Flatten,MaxPooling2D,Dense,Activation,Dropout
```

```python
img.shape
```

```python
model=Sequential()
###create layer--1
model.add(Conv2D(filters=32,kernel_size=(3,3),input_shape=(128,128,3),activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
##crate a layer--2 
model.add(Conv2D(filters=64,kernel_size=(3,3),input_shape=(128,128,3),activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))

##create a layer-3
model.add(Conv2D(filters=64,kernel_size=(3,3),input_shape=(128,128,3),activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))

##convert 2d to 1d
model.add(Flatten())

##dense input dense layer
model.add(Dense(512))
model.add(Activation('relu'))
model.add(Dropout(0.5))

##output dense layer
model.add(Dense(1))
model.add(Activation('sigmoid'))

##compilation 
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.summary()
```

```python
train_img=image_gen.flow_from_directory(path+'/train',batch_size=32,target_size=(128,128),class_mode='binary')
```

```python
validation_img=image_gen.flow_from_directory(path+'/validation',batch_size=32,target_size=(128,128),class_mode='binary')
```

```python
result=model.fit_generator(train_img,
                           steps_per_epoch=163,
                           validation_data=validation_img,
                           validation_steps=17,
                           verbose=1,epochs=100)
```
