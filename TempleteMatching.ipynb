```python
##use a template matching six function 
##using a eval function convert this string to function 
methods=["cv2.TM_CCOEFF","cv2.TM_CCOEFF_NORMED","cv2.TM_CCORR","cv2.TM_CCORR_NORMED","cv2.TM_SQDIFF","cv2.TM_SQDIFF_NORMED"]
```

```python
##use a template matching six function 
##using a eval function convert this string to function 
# methods=['cv2.TM_CCOEFF','cv2.TM_CCOEFF_NORMED','cv2.TM_CCORR','cv2.TM_CCORR_NORMED','cv2.TM_SQDIFF','cv2.TM_SQDIFF_NORMED']
```

```python
##co-ordinate always comes x,y form but shape height and width,so to get bottom corner x+width,y+height
##check template using all method
##if you want to get vaue with position use enumerate function

for m in methods:
    ##create a copy image
    writeGmailcp=writeGmail.copy()
    method=eval(m)
    ##use template matching function
    match=cv2.matchTemplate(writeGmailcp,gmail,method)
    min_val,max_val,min_loc,max_loc=cv2.minMaxLoc(match)
    if method in [cv2.TM_SQDIFF,cv2.TM_CCOEFF_NORMED]:
        top_left=min_loc ###(x,y)
    else:
        top_left=max_loc
    height,width,channels=gmail.shape    
    bottom_right=(top_left[0]+width,top_left[1]+height) ##donot understand
    
    cv2.rectangle(writeGmailcp,top_left,bottom_right,color=(0,255,0),thickness=10)
    plt.subplot(121)
    plt.imshow(match)
    plt.title("Heat of template matching")
    plt.subplot(122)
    plt.imshow(writeGmailcp)
    plt.title("Detection of template")
    plt.suptitle(m)
    plt.show()
    print('\n')
#     print(method)
#     print(m)
    
```
