```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
butterFly=cv2.imread("images/butterfly.png")
butterFly=cv2.cvtColor(butterFly,cv2.COLOR_BGR2RGB)
plt.imshow(butterFly)
```

```python
edge=cv2.Canny(butterFly,threshold1=127,threshold2=127)
plt.imshow(edge,cmap='gray')
```

```python
##calculate median value of images
med_val=np.median(butterFly)
med_val
```

```python
##Lower Threshold value 0 or either 70% of median value
lower=int(max(0,0.7*med_val))
##Upper threshold value 130% of medium value or 255
upper=int(max(255,1.3*med_val))
print('lower:',lower)
print('upper:',upper)
```

```python
edge=cv2.Canny(image=butterFly,threshold1=lower,threshold2=upper)
plt.imshow(edge,cmap='gray')
```

```python
blur_img=cv2.blur(butterFly,ksize=(7,7)) ##when kernel size more the edge are surply show
plt.imshow(blur_img)
```

```python
edge_blur=cv2.Canny(image=blur_img,threshold1=lower,threshold2=upper)
plt.imshow(edge_blur,cmap='gray')
```
