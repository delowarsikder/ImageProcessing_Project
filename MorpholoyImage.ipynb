```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
##create a blank image
##load_image fuction to read a image
def load_img():
    blank_img=np.zeros(shape=(600,600))
    font=cv2.FONT_HERSHEY_COMPLEX
    cv2.putText(blank_img,text="OPENCV",org=(20,300),fontFace=font,fontScale=4,color=(255,0,0),thickness=20)
    return blank_img
```

```python
###display image using create a function
def display_img(img):
    fig=plt.figure(figsize=(5,5))
    ax=fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
```

```python
img=load_img()
# plt.imshow(img)
display_img(img)
```

```python
##create a kernel
kernel=np.ones(shape=(5,5),dtype=np.uint8)
kernel
```

```python
###erosion to extract background feature
result=cv2.erode(img,kernel,iterations=2)
display_img(result)
```

```python
###create a noise
white_noise=np.random.randint(low=0,high=2,size=(600,600))
display_img(white_noise)
```

```python
###convert white nose 0-1 to 0-255
white_noise=white_noise*255
display_img(white_noise)
```

```python
##add white noise with image which is create as blank with a Opencv text
###this adding getting error when image size 800*800
noise_img=white_noise+img
display_img(noise_img)
```

```python
###applying a morphological operation on image
##opening image
opening=cv2.morphologyEx(noise_img,cv2.MORPH_OPEN,kernel)
# plt.imshow(img)
display_img(opening)
```

```python
###create a black noise
black_noise=np.random.randint(low=0,high=2,size=(600,600))
black_noise=black_noise*-255
display_img(black_noise)
```

```python
black_noise_img=black_noise+img
black_noise_img[black_noise_img==-255]=0
display_img(black_noise_img)
```

```python
##closing the blank_noise background 
closing=cv2.morphologyEx(black_noise_img,cv2.MORPH_CLOSE,kernel)
display_img(closing)
```

```python
##edge detection techine using gradiant
gradiant=cv2.morphologyEx(img,cv2.MORPH_GRADIENT,kernel)
display_img(gradiant)
```
