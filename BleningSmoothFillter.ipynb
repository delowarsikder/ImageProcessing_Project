```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
# %matplotlib inline
```

```python
###load image using a function
def load_img():
    img=cv2.imread("images/pond.png").astype(np.float32)/255
    img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    return img
```

```python
# load_img()
```

```python
###display image using large scale but not shape remain unchanged
def display_img(img):
    fig=plt.figure(figsize=(10,20))
    ax=fig.add_subplot(111)
    ax.imshow(img)
```

```python
fix_img=load_img()
display_img(fix_img)
```

```python
##gamma correction
gamma=2
result_gamma=np.power(fix_img,gamma)
display_img(result_gamma)
```

```python
##write text on image
pond_img=load_img()
font=cv2.FONT_HERSHEY_COMPLEX
cv2.putText(pond_img,text="pond",org=(0,250),fontFace=font,fontScale=5,color=(255,0,0),thickness=2)
display_img(pond_img)
```

```python
##desing a kernel
kernel=np.ones(shape=(5,5),dtype=np.float32)/25
# kernel
```

```python
#applying kernel on image
filterImg=cv2.filter2D(pond_img,-1,kernel=kernel)
display_img(filterImg)
```

```python
##disply image again
display_img(pond_img)
```

```python
##build in blur in open cv
blurTest=cv2.blur(pond_img,ksize=(10,10))
display_img(blurTest)
```

```python
##Gaussian blur
gaussianBlur=cv2.GaussianBlur(pond_img,(5,5),5)
plt.imshow(gaussianBlur)
# display_img(gaussianBlur)
```

```python
###median blur
###median blur remove noise form image
mediamBlur_img=cv2.medianBlur(pond_img,5)
plt.imshow(mediamBlur_img)
```

```python
###bilateral Filter 
blur=cv2.bilateralFilter(pond_img,9,75,75)
plt.imshow(blur)
```
