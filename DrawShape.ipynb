```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline




########test back ground
#gmail logo
img=np.zeros(shape=(512,512,3),dtype=np.int16) #create blnk imgae with
img=img+100
cv2.rectangle(img,pt1=(100,100),pt2=(400,300),color=(255,255,255),thickness=-10)
# cv2.rectangle(img,pt1=(150,150),pt2=(350,250),color=(255,255,0),thickness=-10)
#arrow sign
cv2.line(img,pt1=(105,110),pt2=(240,200),color=(0,0,255),thickness=30) 
cv2.line(img,pt1=(240,200),pt2=(390,110),color=(0,0,255),thickness=30)
#side line
cv2.line(img,pt1=(105,110),pt2=(105,290),color=(0,0,255),thickness=30) #left side line
cv2.line(img,pt1=(395,110),pt2=(395,290),color=(0,0,255),thickness=30) #right side line
gmail=img.copy()
cv2.imwrite('images/gmail.png',gmail)
flip=cv2.flip(gmail,1)
plt.imshow(flip)
```

```python
#gmail logo
img=np.zeros(shape=(512,512,3),dtype=np.int16) #create blnk imgae with 
cv2.rectangle(img,pt1=(100,100),pt2=(400,300),color=(255,255,255),thickness=-10)
# cv2.rectangle(img,pt1=(150,150),pt2=(350,250),color=(255,255,0),thickness=-10)
#arrow sign
cv2.line(img,pt1=(105,110),pt2=(240,200),color=(0,0,255),thickness=30) 
cv2.line(img,pt1=(240,200),pt2=(390,110),color=(0,0,255),thickness=30)
#side line
cv2.line(img,pt1=(105,110),pt2=(105,290),color=(0,0,255),thickness=30) #left side line
cv2.line(img,pt1=(395,110),pt2=(395,290),color=(0,0,255),thickness=30) #right side line
gmail=img.copy()
cv2.imwrite('images/gmail.png',gmail)
plt.imshow(img)
```

```python
#write text in image
font=cv2.FONT_HERSHEY_COMPLEX_SMALL
cv2.putText(gmail,text='Gmail',org=(100,400),fontFace=font,fontScale=4,color=(0,0,255),thickness=10,lineType=cv2.LINE_AA)
cv2.imwrite('images/Writegmail.png',gmail)
plt.imshow(gmail)
```
