```python
chessBoard=cv2.imread("images/chessBoard.png",cv2.IMREAD_GRAYSCALE)
plt.imshow(chessBoard,cmap='gray')
```

```python
counter,hierarchy=cv2.findContours(chessBoard,cv2.RETR_CCOMP,cv2.CHAIN_APPROX_SIMPLE)
```

```python
external_couter=np.zeros(chessBoard.shape)
```

```python
for i in range(len(counter)):
    if hierarchy[0][i][3]==-1:
        cv2.drawContours(external_couter,counter,i,255,1)
```

```python
plt.imshow(external_couter,cmap='gray')
```

```python
internal_counter=np.zeros(chessBoard.shape)
```

```python
for i in range(len(counter)):
    if hierarchy[0][i][3]!=-1:
        cv2.drawContours(internal_counter,counter,i,255,1)
```

```python
plt.imshow(internal_counter,cmap='gray')
```
