```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img1=cv2.imread("images/pond.png",cv2.IMREAD_COLOR)
img1=cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
img2=cv2.imread("images/river.png")
img2=cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
```

```python
###reshape image
# plt.imshow(img1)
# img1.shape
img1_resize=cv2.resize(img1,(347,145)) ###resize one processs
plt.imshow(img1_resize)
img1_resize.shape
```

```python
plt.imshow(img2)
img2.shape
```

```python
## Blending two image
####combind two image process-1
alpha=0.8
beta=0.7
gamma=20
blend_img=cv2.addWeighted(img1_resize,alpha,img2,beta,gamma)
plt.imshow(blend_img)
```

```python
#####blending formula img=img1*αlpha+img2*beta+gamma
####combind two image process-2

blend=img1_resize.copy()
row,col,chal=blend.shape
a=.5
b=.5
c=0
for i in range(blend.shape[0]):#row
    for j in range(blend.shape[1]):#col
        for k in range(chal):
            px=img1_resize.item(i,j,k)*a+img2.item(i,j,k)*b+c
            blend.itemset((i,j,k),px)
#             blend[i,j,k]=img1_resize[i,j,k]*a+img2[i,j,k]*b+c ###access px using co-ordinate
plt.imshow(blend)
```

```python
####combind two image process-3
large_img=img1
small_img=img2
x_offset=0
y_offset=0
x_end=x_offset+small_img.shape[0]
y_end=y_offset+small_img.shape[1]
```

```python
###image is 2d array which image.shape[0] return row of array but this is y axis when we consider it ,in co-ordinate system
### and image.shape[1] return column of array but this is x axis when we consider it ,in co-ordinate system

##image[row,col] so, it is also like image[image.shape[0],image.shape[1]]
##some times it seems a spiral concept so it easy to understand if we consider or think link row and col
###(0,0) is started upper left corner and end bottom lower corner

print('x axis',x_end)
print('y axis',y_end)
```

```python
large_img[x_offset:x_end,y_offset:y_end,:]=small_img
plt.imshow(large_img)
```
