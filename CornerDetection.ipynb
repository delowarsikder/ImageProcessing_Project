```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
suduko_img=cv2.imread("images/suduko.png")
suduko_img=cv2.cvtColor(suduko_img,cv2.COLOR_BGR2RGB)
```

```python
suduko_gray_img=cv2.cvtColor(suduko_img,cv2.COLOR_BGR2GRAY)
plt.imshow(suduko_gray_img,cmap='gray')
```

```python
gray=np.float32(suduko_gray_img)
```

```python
###applying a herris corner detection algorithm
dst=cv2.cornerHarris(src=gray,blockSize=2,ksize=3,k=0.04)
```

```python
dst=cv2.dilate(dst,None)
```

```python
suduko_img[dst>0.01*dst.max()]=[255,0,0]
```

```python
plt.imshow(suduko_img)
```

```python
suduko_img=cv2.imread("images/suduko.png")
suduko_img=cv2.cvtColor(suduko_img,cv2.COLOR_BGR2RGB)
```

```python
suduko_gray_img=cv2.cvtColor(suduko_img,cv2.COLOR_BGR2GRAY)
plt.imshow(suduko_gray_img,cmap='gray')
```

```python
corner=cv2.goodFeaturesToTrack(suduko_gray_img,81,0.01,10)
```

```python
corner=np.int0(corner)
```

```python
for i in corner:
    x,y=i.ravel()
    cv2.circle(suduko_img,(x,y),3,(255,0,0),-1)
```

```python
plt.imshow(suduko_img)
```
