```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img=cv2.imread('images/pond.png')
plt.imshow(img)
# img.shape
```

```python
img_rgb=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
##HSV --hue ,saturation and value
img_hsv=cv2.cvtColor(img_rgb,cv2.COLOR_RGB2HSV)
plt.imshow(img_hsv)
```

```python
#######HLS ---huelightness and saturaion
img_hls=cv2.cvtColor(img_rgb,cv2.COLOR_RGB2HLS)
plt.imshow(img_hls)
```
